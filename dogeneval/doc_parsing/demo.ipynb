```python
import os


folder = 'sampled_pages_200_txt'
filename_list = []
for f in os.listdir(folder):
        # 构建完整的文件路径
        f = f.rsplit('.', 1)[0]
        filename_list.append(f)
```

```python
def clean_initial_section(input_file_path, output_file_path):
    with open(input_file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
    if lines and (lines[0].strip().startswith('##') or not lines[2].strip().startswith('*')) :
        # 如果以##开头，只删除加粗标记的内容
        with open(output_file_path, 'w', encoding='utf-8') as file:
            for line in lines:
                cleaned_line = line.replace('**', '')
                file.write(cleaned_line)
    else:
        with open(output_file_path, 'w', encoding='utf-8') as file:
            skip = False
            flag = False
            for line in lines:
                if skip:
                    if line.strip().startswith('#'):
                        skip = False
                        flag = True
                        file.write(line)
                        continue
                        
                if line.strip().startswith('#'):
                    file.write(line)
                elif line.strip().startswith('*'):
                    if flag!=True:
                        skip = True
                    else:
                        file.write(line)
                else:
                    file.write(line)
        with open(output_file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()

        with open(output_file_path, 'w', encoding='utf-8') as file:
            for line in lines:
                # 替换掉两个*加粗标记
                cleaned_line = line.replace('**', '')
                file.write(cleaned_line)
# 使用这个函数，传入你的输入文件和输出文件路径
folder = 'sampled_pages_200_txt'
for filename in filename_list:
    clean_initial_section(folder+'/'+filename + '.txt', 'tmp'+'/'+filename + '_tmp.txt')
```
